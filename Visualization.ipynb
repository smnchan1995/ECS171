# JOKERater Database

````python
import sqlite3
import pandas as pd
from sqlalchemy import text

con = sqlite3.connect("joke.sqlite3")
df = pd.read_sql_query("SELECT * FROM JokeRater", con)
# verify that result of SQL query is stored in the dataframe
con.close()

df[0:5]
````

### Who are the Raters?

````python
import matplotlib.pyplot as plt
%matplotlib inline
# By gender
gender_df = pd.to_numeric(df['gender'])
gender_df.plot.pie()
````
